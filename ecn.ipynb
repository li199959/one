```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

# Replace with the path of your uploaded file
file_path = '/content/原始数据 .xlsx'

# Load the data
data = pd.read_excel(file_path)

# Create an empty figure
fig = go.Figure()

# Loop over each column and add a trace to the figure for the CDF
for i, column in enumerate(data.columns):
    trace_name = column

    # Sort the column data and calculate the CDF
    sorted_data = np.sort(data[column])
    y = np.arange(1, len(sorted_data) + 1) / len(sorted_data)

    line_type = 'solid'
    if i == len(data.columns) - 1:  # If it is the last column, set line type to dash
        line_type = 'dash'

    # Add trace to the figure with updated name and axes
    fig.add_trace(go.Scatter(x=y, y=sorted_data, mode='lines', name=trace_name, line=dict(dash=line_type)))

# Define the buttons
num_columns = len(data.columns)
last_column_index = num_columns - 1

buttons = [
    dict(label='Select All', method='update', args=[{'visible': [True] * num_columns}])  # "Select All" button
]

for i in range(4):
    start = i * 10
    end = start + 10
    if end >= num_columns:
        end = num_columns - 1  # Exclude the last column

    # Set visibility to True for columns in the range [start, end] and the last column
    visibility_array = [start <= j < end or j == last_column_index for j in range(num_columns)]
    buttons.append(dict(label=f'{start}-{end} & Last', method='update', args=[{'visible': visibility_array}]))

# Update layout with buttons for selective display
fig.update_layout(
    title='Cumulative Distribution Function (CDF)',
    xaxis_title='Cumulative Probability',
    xaxis=dict(range=[0, 1]),
    yaxis_title='Value',
    font=dict(size=10),
    hovermode='closest',
    updatemenus=[dict(x=1.15, y=1, yanchor='top', xanchor='left', buttons=buttons)]
)

# Show the interactive plot in Colab
fig.show()

# Save the interactive plot as an HTML file
html_file_path = '/content/ecn_cdf_plot_with_grouped_selection_and_dash_last_column.html'
fig.write_html(html_file_path)

print(f"The plot has been saved to {html_file_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 请确保您的文件路径是正确的
file_path = '您的文件路径'  # 例如: '/content/4组.xlsx'

# 读取 Excel 文件
xls = pd.ExcelFile(file_path)

# 读取第一个工作表的内容
sheet1_df = xls.parse('Sheet1')

# 计算每一列的均值与方差
mean_variance = pd.DataFrame(index=['mean', 'variance'])
for column in sheet1_df.columns:
    mean_variance[column] = [sheet1_df[column].mean(), sheet1_df[column].var()]

# 打印均值与方差
print(mean_variance)

# 设置均值 y 轴的范围
mean_y_axis_min = 29.5  # 您可以修改这个值
mean_y_axis_max = 30.5  # 您可以修改这个值

# 设置图形和轴
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))
fig.suptitle('Mean and Variance of Each Column', fontsize=16)

# 绘制均值的柱形图
mean_variance.loc['mean'].plot(kind='bar', ax=axs[0], color='skyblue', edgecolor='black')
axs[0].set_title('Mean')
axs[0].set_ylabel('Value')
axs[0].set_ylim(bottom=mean_y_axis_min, top=mean_y_axis_max)  # 设置 y 轴的范围

# 绘制方差的柱形图
mean_variance.loc['variance'].plot(kind='bar', ax=axs[1], color='lightcoral', edgecolor='black')
axs[1].set_title('Variance')
axs[1].set_ylabel('Value')

# 旋转 x 轴标签以便更好地阅读
axs[0].tick_params(axis='x', rotation=45)
axs[1].tick_params(axis='x', rotation=45)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

```
