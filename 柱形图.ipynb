```r
# 加载所需库
library(readxl)

# 读取Excel文件
data <- read_excel("/content/extracted_data.xlsx")

# 只对数值型数据列绘制箱型图
numeric_columns <- sapply(data, is.numeric)
numeric_data <- data[, numeric_columns]

# 检查是否有数值型列
if(sum(numeric_columns) > 0) {
  # 计算子图的行数和列数，这里我们设定每行最多显示2个图
  ncol_sub <- min(2, sum(numeric_columns))
  nrow_sub <- ceiling(sum(numeric_columns) / ncol_sub)

  # 设置子图布局
  par(mfrow=c(nrow_sub, ncol_sub))

  # 对每一列绘制箱型图
  for(i in 1:ncol(numeric_data)) {
    boxplot(numeric_data[,i], main=colnames(numeric_data)[i])
  }
} else {
  warning("No numeric columns found")
}

```
