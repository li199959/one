```python
import pandas as pd
import torch
from transformers import BertTokenizer, BertModel

# Check if CUDA (GPU support) is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load the data
data_path = '/content/grouped_anomalous_content.csv'  # <-- 修改为您的文件路径
data = pd.read_csv(data_path)

# Load pre-trained model and tokenizer, and move model to the device
model_name = 'bert-base-uncased'
model = BertModel.from_pretrained(model_name).to(device)
tokenizer = BertTokenizer.from_pretrained(model_name)

def encode_text(text):
    """
    Encode the text using BERT and return the CLS token embedding.
    """
    # Tokenize and encode the text
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=512).to(device)

    # Get BERT embeddings
    with torch.no_grad():
        outputs = model(**inputs)

    # Extract the [CLS] embedding and move it back to CPU
    cls_embedding = outputs.last_hidden_state[:, 0, :].cpu().numpy()

    return cls_embedding

# Encode all contents
encoded_contents = data['content'].apply(encode_text)

# Now, `encoded_contents` contains the BERT embeddings for each content in the file.

```

```python
# Convert the encoded contents to a DataFrame
encoded_df = pd.DataFrame(list(encoded_contents.apply(lambda x: x[0])))

# Save the encoded vectors to a CSV file
encoded_file_path = 'encoded_vectors.csv'  # <-- 修改为您希望保存文件的路径
encoded_df.to_csv(encoded_file_path, index=False)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score

# Load the encoded BERT vectors
encoded_df = pd.read_csv('/content/encoded_vectors.csv')  # <-- 修改为您的BERT向量文件路径

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(encoded_df)

# 1. K-Means Clustering
# For demonstration, let's assume we want to try with 5 clusters. You might want to determine the best number of clusters using the Elbow method or Silhouette analysis.
kmeans = KMeans(n_clusters=15, random_state=42)
kmeans_labels = kmeans.fit_predict(scaled_data)

# 2. DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(scaled_data)

# 3. Hierarchical Clustering (Agglomerative Clustering)
# Again, for demonstration, let's assume we want to try with 5 clusters.
agg_clustering = AgglomerativeClustering(n_clusters=5, linkage='ward')
agg_labels = agg_clustering.fit_predict(scaled_data)

# Evaluation using Silhouette Score (Higher is better. Range: [-1, 1])
kmeans_silhouette = silhouette_score(scaled_data, kmeans_labels)
dbscan_silhouette = silhouette_score(scaled_data, dbscan_labels) if len(np.unique(dbscan_labels)) > 1 else np.nan  # DBSCAN might produce a single cluster which can't be evaluated using silhouette score
agg_silhouette = silhouette_score(scaled_data, agg_labels)

print(f"Silhouette Score for K-Means: {kmeans_silhouette}")
print(f"Silhouette Score for DBSCAN: {dbscan_silhouette}")
print(f"Silhouette Score for Agglomerative Clustering: {agg_silhouette}")
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Define a range for the potential number of clusters
k_values = range(2, 15)  # Assuming you want to check for up to 15 clusters. Adjust as needed.

wcss = []  # For the elbow method
silhouettes = []  # For the silhouette score

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans_labels = kmeans.fit_predict(scaled_data)

    # For the elbow method
    wcss.append(kmeans.inertia_)

    # For the silhouette score
    silhouettes.append(silhouette_score(scaled_data, kmeans_labels))

# Plotting the results

# Elbow method
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(k_values, wcss, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method')

# Silhouette score
plt.subplot(1, 2, 2)
plt.plot(k_values, silhouettes, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Analysis')

plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Define a range for the potential number of clusters
k_values = range(2, 15)  # Assuming you want to check for up to 15 clusters. Adjust as needed.

wcss = []  # For the elbow method
silhouettes = []  # For the silhouette score

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans_labels = kmeans.fit_predict(scaled_data)

    # For the elbow method
    wcss.append(kmeans.inertia_)

    # For the silhouette score
    silhouettes.append(silhouette_score(scaled_data, kmeans_labels))

# Determine the "elbow" from the WCSS values
wcss_diff = np.diff(wcss)
elbow_k = np.argmin(wcss_diff) + 1  # +1 because diff is 1 element shorter

# Determine the optimal number of clusters from the silhouette scores
optimal_k_silhouette = k_values[np.argmax(silhouettes)]

# Print the optimal k values
print(f"Optimal k value based on the Elbow method: {elbow_k}")
print(f"Optimal k value based on the Silhouette Score: {optimal_k_silhouette}")

# Plotting the results

# Elbow method
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(k_values, wcss, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method')

# Silhouette score
plt.subplot(1, 2, 2)
plt.plot(k_values, silhouettes, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Analysis')

plt.tight_layout()
plt.show()
```
