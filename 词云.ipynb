```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# 
with open('.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# 
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# 
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import jieba

# 
with open('.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# jieba
word_list = jieba.cut(text)
processed_text = ' '.join(word_list)

# 
wordcloud = WordCloud(width=2000, height=1000, background_color='white', font_path='font.ttf').generate(processed_text)

# 
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import jieba
from PIL import Image
import numpy as np

# 
with open('.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# jieba
word_list = jieba.cut(text)
processed_text = ' '.join(word_list)

# 
mask = np.array(Image.open('1234.png'))

# 
wordcloud = WordCloud(width=1000, height=500, background_color='white', font_path='font.ttf', mask=mask,
                      collocations=False, max_font_size=50, min_font_size=10, max_words=200).generate(processed_text)

# 
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

```
